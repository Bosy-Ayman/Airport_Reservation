```python
import numpy as np
import pandas as pd #data structure
```

```python
df_can = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')
```

```python
df_can.head()
```

```python
df_can.set_index('Country',inplace=True) #Transform the countries into the main index
```

```python
df_can.head(3)
```

```python
df_can.index.name = None #remove the name of the index
```

```python
df_can.head()
```

```python
years = list(map(str, range(1980, 2014)))
years
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
haiti = df_can.loc['Haiti', years] # passing in years 1980 - 2013 to exclude the 'total' column
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.plot(kind ='line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# annotate the 2010 Earthquake.
# syntax: plt.text(x, y, label)
plt.text(2000, 6000, '2010 Earthquake') # see note below

plt.show()
```

```python
#Step 1: Get the data set for China and India, and display the dataframe.
df_CI = df_can.loc[['India', 'China'], years]
df_CI = df_CI.transpose() #???
```

```python
df_CI.index = df_CI.index.map(int)
df_CI.plot(kind='line')
plt.title("Imigration from China and India")
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
inplace = True
df_can.sort_values(by = 'Total',ascending = False,axis = 0,inplace = True)
```
